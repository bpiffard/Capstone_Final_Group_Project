```python
import pandas as pd
```

```python
# Importing the data
file_path = 'Resources/all_data.csv'
df = pd.read_csv(file_path)
```

```python
# Converting df dtypes
columns_to_convert = ['store_sqft', 'grocery_sqft',
       'frozen_sqft', 'meat_sqft', 'coffee_bar', 'video_store', 'salad_bar',
       'prepared_food', 'florist','unit_sales_in_millions', 'total_children',
        'avg_cars_at_home_approx', 'num_children_at_home', 'recyclable_package', 'low_fat',
        'units_per_case']
for column in columns_to_convert:
    df = df.astype({column:int})
```

```python
# Making store df
store_col = ['store_type', 'store_sqft', 'grocery_sqft', 'frozen_sqft', 'meat_sqft',
    'coffee_bar', 'video_store', 'salad_bar', 'prepared_food', 'florist', 'store_city', 'store_state']
store_df = df[store_col].drop_duplicates()
```

```python
# Creating id column for store df
store_df['store_id'] = ''
id_num = store_df.index
store_id = []
for num in id_num:
    store_id.append(f's{num}')

store_df['store_id'] = store_id
store_df.head()
```

```python
# Making location table
loc_col =['store_city', 'store_state']
location_df = df[loc_col].drop_duplicates()
```

```python
# Making location id
location_df['location_id'] = ''
id_num = location_df.index
loc_id = []
for num in id_num:
    loc_id.append(f'loc{num}')
    
location_df['location_id'] = loc_id
location_df.head()
```

```python
# Creating customer df
cust_col= ['marital_status', 'gender', 'total_children', 'education', 'member_card',
                  'occupation', 'houseowner', 'avg_cars_at_home_approx', 'avg_yearly_income',
                  'num_children_at_home', 'store_city', 'store_state']
customer_df = df[cust_col].drop_duplicates()
```

```python
# Creating customer id
customer_df['customer_id'] = ''
id_num = customer_df.index
customer_id = []
for num in id_num:
    customer_id.append(f'c{num}')
    
customer_df['customer_id'] = customer_id
customer_df.head()
```

```python
# Creating product df
product_col = ['brand_name', 'food_category', 'food_department', 'food_family',
                 'SRP', 'gross_weight', 'net_weight', 'recyclable_package', 'low_fat',
                 'units_per_case']
product_df = df[product_col].drop_duplicates()
```

```python
# Creating product id
product_df['product_id'] = ''
id_num = product_df.index
prod_id = []
for num in id_num:
    prod_id.append(f'prod{num}')
    
product_df['product_id'] = prod_id
product_df.head()
```

```python
# Making function to add keys to primary table
def join_tables(key_df, cols):
    global promo_df
    promo_df = promo_df.merge(
        key_df,
        how='right',
        on=cols
    )
    return promo_df
```

```python
# Adding the keys!
promo_df = df.copy()
join_tables(store_df,store_col)
join_tables(customer_df,cust_col)
join_tables(product_df,product_col)
join_tables(location_df,loc_col)
promo_df.columns
```

```python
# Creating promo_id column
promo_df['promotion_id'] = ''
id_num = promo_df.index
promo_id = []
for num in id_num:
    promo_id.append(f'prom{num}')
    
promo_df['promotion_id'] = promo_id
promo_df.head()
```

```python
# Dropping and Rearragning columns in promo_df
promo_df = promo_df[['promotion_id', 'promotion_name', 'media_type', 'cost',
    'store_sales_in_millions', 'store_cost_in_millions', 'unit_sales_in_millions',
    'product_id', 'store_id', 'customer_id', 'location_id']]
promo_df.columns
```

```python
# Adding the location keys to store_data
store_df = store_df.merge(
    location_df,
    how='left',
    on=loc_col
)
store_df.drop(columns=loc_col,inplace=True)
store_df.columns
```

```python
# Adding the location keys to customer_data
customer_df = customer_df.merge(
    location_df,
    how='left',
    on=loc_col
)
customer_df.drop(columns=loc_col,inplace=True)
customer_df.columns
```

```python
location_df.columns
```

```python
# Rearranging columns in tables
store_df = store_df[['store_id', 'store_type', 'store_sqft', 'grocery_sqft', 'frozen_sqft', 'meat_sqft',
       'coffee_bar', 'video_store', 'salad_bar', 'prepared_food', 'florist','location_id']]
customer_df = customer_df[['customer_id','marital_status', 'gender', 'total_children', 'education',
       'member_card', 'occupation', 'houseowner', 'avg_cars_at_home_approx',
       'avg_yearly_income', 'num_children_at_home','location_id']]
product_df = product_df[['product_id', 'brand_name', 'food_category', 'food_department', 'food_family',
        'SRP','gross_weight', 'net_weight', 'recyclable_package', 'low_fat', 'units_per_case']]
location_df = location_df[['location_id', 'store_city', 'store_state']]
```

```python
# Exporting the data to .csv
file_path = 'New_ERD_Tables/'
promo_df.to_csv(f'{file_path}promotion_data.csv', index=False)
store_df.to_csv(f'{file_path}store_data.csv', index=False)
customer_df.to_csv(f'{file_path}customer_data.csv', index=False)
product_df.to_csv(f'{file_path}product_data.csv', index=False)
location_df.to_csv(f'{file_path}location_data.csv', index=False)
```
