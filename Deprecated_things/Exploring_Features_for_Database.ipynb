```python
import pandas as pd
```

```python
# Importing the data
file_path = '../Resources/all_data.csv'
df = pd.read_csv(file_path)
```

```python
# Creating desired shape for promotion data
promo_df = df[['promotion_name', 'media_type', 'cost', 'store_sales_in_millions',
               'store_cost_in_millions', 'unit_sales_in_millions']].drop_duplicates()
promo_df.head()
```

```python
# Seeing which duplicate data was lost
missing_data = df.index.difference(promo_df.index)
missing_data
```

```python
df.iloc[missing_data]
```

```python
# Looking at datatypes
promo_df.dtypes
```

```python
# Converting datatypes
promo_df = promo_df.astype({'unit_sales_in_millions':int})
promo_df.dtypes
```

```python
# Creating desired shape for location data
location_df = df[['store_city', 'store_state']].drop_duplicates()
location_df.head()
```

```python
# Converting store df to desired shape
store_df = df[['store_type', 'store_sqft', 'grocery_sqft', 'frozen_sqft', 'meat_sqft',
               'coffee_bar', 'video_store', 'salad_bar', 'prepared_food', 'florist']].drop_duplicates()
store_df.head()
```

```python
# Converting store df dtypes
columns_to_convert = ['store_sqft', 'grocery_sqft',
       'frozen_sqft', 'meat_sqft', 'coffee_bar', 'video_store', 'salad_bar',
       'prepared_food', 'florist']
for column in columns_to_convert:
    store_df = store_df.astype({column:int})
store_df.dtypes
```

```python
# Same thing for the customer profile data
customer_df = df[['marital_status', 'gender', 'total_children', 'education', 'member_card',
                  'occupation', 'houseowner', 'avg_cars_at_home_approx', 'avg_yearly_income',
                  'num_children_at_home']].drop_duplicates()
customer_df.head()
```

```python
customer_df.dtypes
```

```python
# Fixing dtypes
columns_to_convert = ['total_children', 'avg_cars_at_home_approx', 'num_children_at_home']
for column in columns_to_convert:
    customer_df = customer_df.astype({column:int})
customer_df.dtypes
```

```python
# Doing the same for product data...
product_df = df[['brand_name', 'food_category', 'food_department', 'food_family',
                 'SRP', 'gross_weight', 'net_weight', 'recyclable_package', 'low_fat',
                 'units_per_case']].drop_duplicates()
product_df.head()
```

```python
# Checking if brand_name can be used as a key? But then that would look weird on the promo table maybe..
print(len(product_df))
print(product_df['brand_name'].unique)
```

```python
product_df.dtypes
```

```python
columns_to_convert = ['recyclable_package', 'low_fat', 'units_per_case']
for column in columns_to_convert:
    product_df = product_df.astype({column:int})
product_df.dtypes
```

```python
# Exporting the dfs to csv files
promo_df.to_csv('ERD_Resources/promotion_data.csv', index=False)
location_df.to_csv('ERD_Resources/location_data.csv', index=False)
store_df.to_csv('ERD_Resources/store_data.csv', index=False)
customer_df.to_csv('ERD_Resources/customer_data.csv', index=False)
product_df.to_csv('ERD_Resources/product_data.csv', index=False)
```
