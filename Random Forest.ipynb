```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import sklearn.metrics as metrics
from sklearn.ensemble import RandomForestClassifier
```

```python
# Importing the data
file_path = 'Resources/store_data.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
le = LabelEncoder()
duplicate_df = df.copy()

encode_columns = ['food_category', 'food_family','food_department', 'promotion_name', 'brand_name',
                    'store_type','store_city', 'store_state']

for label in encode_columns:
    duplicate_df[label] = le.fit_transform(duplicate_df[label])

duplicate_df = duplicate_df.drop(columns=['id','media'])
duplicate_df.head()
```
