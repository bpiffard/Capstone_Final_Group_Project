```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf
from path import Path
```

```python
file_path = 'Resources/store_data.csv'
store_df = pd.read_csv(file_path)
store_df.head()
```

```python
store_df.drop(['sales_country','promotion_name','food_category'], axis=1, inplace=True)
store_df.head()
```

```python
food_family = store_df.dtypes[store_df.dtypes =="object"].index.tolist()
```

```python
food_family
```

```python
# 8 Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(store_df[food_family]))

# Add the encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names(food_family)
encode_df.head(10)
```

```python
store_df = store_df.merge(encode_df, left_index=True, right_index=True)
store_df.head()
```

```python
store_df = store_df.drop(columns = food_family)
store_df.head()
```

```python
store_df=store_df.rename(columns={"unit_sales(in millions)": "unit_sales"})
store_df.head()
```

```python
#Split our preprocessed data into our features and target arrays
y = store_df.media_type_TV
X = store_df.drop(columns= "media_type_TV")

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
 # Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train_scaled[0])

nn = tf.keras.models.Sequential()

# 3 First hidden layer
nn.add(tf.keras.layers.Dense(units=800, activation="sigmoid", input_dim = number_input_features))

# 4 Second hidden layer
nn.add(tf.keras.layers.Dense(units=500, activation="relu"))

# 5 Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="linear"))

# 6 Check the structure of the model
nn.summary()
```

```python
#  Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=['accuracy'])
```

```python
#  Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=5)
```

```python
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
